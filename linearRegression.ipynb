### model:
$$
f_{w,b}(x)=wx+b
$$

### parameters:
$$
w,b
$$

### Cost function:Squared error cost function
$$
J(w,b)=
\frac{1}{2m}\sum_{i=1}^{m}(\widehat{y}_i-y_i)^2\\
m=number\; of\; training\; examples\\
J(w,b)=
\frac{1}{2m}\sum_{i=1}^{m}(f_{w,b}(x_i)-y_i)^2
$$

### goal:
$$
\underset{w,b}{minimize}\,J(w,b)
$$

```python
import numpy as np
x = np.array(list(range(10)))
y = np.array([(2*i+5) for i in x])+np.random.normal(size=10)
y
```

```python
import matplotlib.pyplot as plt
y_t = np.array([(2*i+5) for i in x])
plt.plot(x,y,'r*')
plt.plot(x,y_t)
plt.show()
```

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
w_i = list(range(-50,50))
b_i = list(range(-50,50))
w_draw = list()
b_draw = list()
cost_draw = list()
m=10
for w in w_i:
    for b in b_i:
        w_draw.append(w)
        b_draw.append(b)
        cost = (1/2/m)*sum(((w*x[i]+b-y[i])**2) for i in range(10))
        cost_draw.append(cost)
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(w_draw, b_draw, cost_draw,
                vmin=min(cost_draw) * 2, cmap=cm.Blues)
plt.show()
```

```python
from ml.regression import linerRegression
model = linerRegression.LinerRegression(x, y)
w, b = model.gradient_descent(0.01, 2500)
y_p = [(x[i]*w+b) for i in range(10)]
print("w:", w, "b:", b)
plt.plot(x, y, 'r*')
plt.plot(x, y_p)
plt.show()
```

```python
history = np.array(model.get_history())
cost = history[:, -1]
print("cost", model.get_cost())
times = np.array(range(len(cost)))
plt.plot(times, cost)
plt.show()
```
